# Model building

**Here the problem is classification statement so we are using decision tree classifier**

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,:-1]
y = df.target
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=54)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred_test = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred_test)
```

**Let us visualize the tree we got here**

```python
from sklearn import tree
plt.figure(figsize=(35,25))
tree.plot_tree(model,filled=True)
```

**Still we can get more accuracy by pruning technique**

**Minimal cost-complexity pruning is an algorithm used to prune a tree to avoid over-fitting. This pruning technique is parameterized by the cost complexity parameter, ccp_alpha. Greater values of ccp_alpha increase the number of nodes pruned.**

```python
path = model.cost_complexity_pruning_path(x_train,y_train)
```

```python
ccp_alphas,impurities = path.ccp_alphas,path.impurities
```

```python
ccp_alphas
```

```python
pred_clf = []
for i in ccp_alphas:
    model = DecisionTreeClassifier(ccp_alpha=i)
    model.fit(x_train,y_train)
    pred_clf.append(model)
```

```python
train_score = [i.score(x_train,y_train) for i in pred_clf]
test_score = [i.score(x_test,y_test) for i in pred_clf]
```

```python
fig = plt.figure(figsize=(20,5))
ax = fig.add_axes([0,0,1,1])
ax.plot(ccp_alphas,train_score,drawstyle="steps-post",marker='o')
ax.set_xlabel('ccp_alpha')
ax.set_ylabel('train_score')
ax.plot(ccp_alphas,test_score,drawstyle="steps-post",marker='o')
ax.set_xlabel('ccp_alpha')
ax.set_ylabel('test_score')
ax.set_xticks([0.01,0.02,0.03,0.04])
ax.legend(['train','test'])
plt.show()
```

**We can choose ccp_alpha=0.15**

```python
model = DecisionTreeClassifier(ccp_alpha=0.02)
model.fit(x_train,y_train)
y_pred_test1 = model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_test1)
```

**Before pruning we got accuracy as 0.935672514619883**

**After pruning we got accuracy as 0.9415204678362573**
